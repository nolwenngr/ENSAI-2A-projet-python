Régression linéaire pour expliquer les variations de fréquentations

```python
from sklearn.linear_model import LinearRegression
# 1. NETTOYAGE (Optionnel mais fortement recommandé)
# On s'assure qu'il n'y a pas de valeurs manquantes dans nos colonnes cibles
df_totale_sans_na = df_freq_totale.dropna(subset=['freq_net', 'NOMREG','annee']).copy()
variable_exp_freq_totale = df_totale_sans_na[ 'NOMREG','annee']
variable_int_freq_totale = df_totale_sans_na[ 'freq_net']

model = LinearRegression()
model.fit(variable_exp_freq_totale, variable_int_freq_totale)

# Afficher les coefficients
print("Coefficient :", model.coef_)
print("Intercept :", model.intercept_)
```
